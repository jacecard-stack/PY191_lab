```python
#second assignment - have each bar with different color, each point labeled(for hour and value), identify minimum and maximum for each line, needs title, y level, x level. Lines should be determined with the mean of the temperatures, the mean is for each day so each day will have an average value. so there will be 9 total bars, 1 for each day for each station(rise-01, rise-02, rise-03)
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from IPython.display import HTML
HTML("<style>.container{width:100%!important;margin:auto}div.cell,div.input_area{padding:2px;margin:0}div.output_wrapper{padding:0;margin:0}</style>")
station_df = pd.read_csv("station_weather.csv")
def avg(program, day):
    RISE = station_df[(station_df["station_id"] == program) & (station_df["date"] == day)]
    mean = RISE["temperature_f"].mean()
    return mean
station1 = [avg("RISE-01", "2026-06-15"), avg("RISE-02", "2026-06-15"), avg("RISE-03", "2026-06-15")]
station2 = [avg("RISE-01", "2026-06-16"), avg("RISE-02", "2026-06-16"), avg("RISE-03", "2026-06-16")]
station3 = [avg("RISE-01", "2026-06-17"), avg("RISE-02", "2026-06-17"), avg("RISE-03", "2026-06-17")]

categories = ['RISE-01', 'RISE-02', 'RISE-03']

x_indexes = np.arange(len(categories))
bar_width = 0.20  # Set an explicit width for each bar chunk

fig, ax = plt.subplots(figsize=(9, 5.5))

bars1 = ax.bar(x_indexes - bar_width, station1, width=bar_width,
               color='#1f77b4', edgecolor='#134a70', linewidth=1.2, label='2026-06-15')

bars2 = ax.bar(x_indexes, station2, width=bar_width,
               color='#ff7f0e', edgecolor='#a85207', linewidth=1.2, label='2026-06-16')

bars3 = ax.bar(x_indexes + bar_width, station3, width=bar_width,
               color='#fb3f0e', edgecolor='#134a70', linewidth=1.2, label='2026-06-17')
ax.set_xticks(x_indexes)
ax.set_xticklabels(categories)

for b1, b2, b3 in zip(bars1, bars2, bars3):
    ax.text(b1.get_x() + b1.get_width()/2, b1.get_height() + 0.4, f"{b1.get_height()}", ha='center', fontsize=9)
    ax.text(b2.get_x() + b2.get_width()/2, b2.get_height() + 0.4, f"{b2.get_height()}", ha='center', fontsize=9)
    ax.text(b3.get_x() + b3.get_width()/2, b3.get_height() + 0.4, f"{b3.get_height()}", ha='center', fontsize=9)

ax.set_title("Mean temperature for each day", fontsize=13, fontweight='bold', pad=15)
ax.set_xlabel("Station", fontsize=11)
ax.set_ylabel("Temperature", fontsize=11)
ax.set_ylim(55, 75)
ax.legend(loc='upper right', frameon=True, shadow=True)


ax.annotate('max avg temperature',
             xy=(14, 78.83333333333333),            # The actual target point coordinates
             xytext=(15.5, 80.5),         # Where the text box will physically sit
             arrowprops=dict(facecolor='black', shrink=0.08, width=1, headwidth=6))

ax.annotate('min avg temperature',
             xy=(8, 52.86666666666667),            # The actual target point coordinates
             xytext=(9, 55),         # Where the text box will physically sit
             arrowprops=dict(facecolor='black', shrink=0.08, width=1, headwidth=6))

plt.tight_layout()
plt.show()

print("June 15")
print("Min:", min(station1))
print("Max:", max(station1))

print("June 16")
print("Min:", min(station2))
print("Max:", max(station2))

print("June 17")
print("Min:", min(station3))
print("Max:", max(station3))


#print(station_df["temperature_f"])
#print(station_df["hour"])
#print(station_df.loc[0:7,["hour", "temperature_f"]])
#RISE1DAY1 = station_df[(station_df["station_id"] == "RISE-01") & (station_df["date"] == "2026-06-15")]
#print(RISE1)
#Day1 = RISE1[RISE1["date"] == "2026-06-15"]
#print(RISE1DAY1)
#print("Mean:", RISE1.loc[0:7, "temperature_f"].mean())
#print("Mean:", RISE1DAY1["temperature_f"].mean())
```

```python
#first assignment - plot line from dataset: Line should be determined with the mean of the temperatures, the mean is taken for each hour(so each day's value for each hour will be averaged using the values from day 1, 2, and 3)) (x axis is hours). Do for each station.

import os
import pandas as pd
import yfinance as yf
import time
import matplotlib.pyplot as plt
from IPython.display import HTML
HTML("<style>.container{width:100%!important;margin:auto}div.cell,div.input_area{padding:2px;margin:0}div.output_wrapper{padding:0;margin:0}</style>")
def avg(hour, id):
    RISE = station_df[(station_df["hour"] == hour) & (station_df["station_id"] == id)]
    mean = RISE["temperature_f"].mean()
    return mean
rise1 = []
rise2 = []
rise3 = []
hours = []
for i in range(6, 22, 2):
    hours.append(i)
    rise1.append(avg(i, "RISE-01"))
    rise2.append(avg(i, "RISE-02"))
    rise3.append(avg(i, "RISE-03"))



fig, ax1 = plt.subplots(figsize=(10, 6))

# Line 1 from before
ax1.plot(hours, rise1, color='#1f77b4', linestyle='-', marker='o', linewidth=2.5, label='rise1')

# Addition: Line 2 with separate tracking styles
ax1.plot(hours, rise2, color='#d62728', linestyle='--', marker='s', linewidth=2.5, label='rise2')

ax1.plot(hours, rise3, color='#134a70', linestyle='--', marker='s', linewidth=2.5, label='rise3')
ax1.set_xlabel("Hour")
ax1.set_ylabel("Average Temperature (°F)")
ax1.set_title("Average Temperature by Hour for Each Station")
plt.legend(loc='upper right', frameon=True, shadow=True)
plt.show()


```
